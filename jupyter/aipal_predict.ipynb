```python
import xgboost as xgb
import pandas as pd
import numpy as np

# Load your model
model_path = 'model_v2.json'
model = xgb.Booster()
model.load_model(model_path)

# Example function to prepare data* and make predictions
def prepare_data_and_predict(age, MCV_fL, MCHC_g_L, PT, WBC_G_L, Lymphocytes_G_L, Monocytes_G_L, Platelets_G_L, fibri_gL, LDH_UI_L):
    # Calculate Monocytes percentage
    mono_percent = (Monocytes_G_L * 100) / WBC_G_L
    
    # Prepare your data frame (or matrix)
    sample_data = pd.DataFrame({
        'fibri_gL': [fibri_gL], 
        'MCV_fL': [MCV_fL], 
        'mono_percent': [mono_percent], 
        'LDH_UI_L': [LDH_UI_L], 
        'PT': [PT], 
        'MCHC_g_L': [MCHC_g_L], 
        'Lymphocytes_G_L': [Lymphocytes_G_L], 
        'age': [age], 
        'Monocytes_G_L': [Monocytes_G_L], 
        'Platelets_G_L': [Platelets_G_L]
    })

    # Convert the data frame to DMatrix
    dtest = xgb.DMatrix(sample_data)

    # Make the prediction
    prediction = model.predict(dtest)
    
    # Print or return the prediction
    print("Prediction:", prediction)
    return prediction

# Call the function with the input values
age = 53
MCV_fL = 90.2
MCHC_g_L = 330
PT = 50
WBC_G_L = 10
Lymphocytes_G_L = 1
Monocytes_G_L = 6
Platelets_G_L = 6
fibri_gL = 6
LDH_UI_L = 250

prepare_data_and_predict(age, MCV_fL, MCHC_g_L, PT, WBC_G_L, Lymphocytes_G_L, Monocytes_G_L, Platelets_G_L, fibri_gL, LDH_UI_L)

```
